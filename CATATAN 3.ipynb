# POLYMORPHISME
Merupakan konsep yang memungkinkan digunakannya suatu interface yang sama untuk memerintah suatu object agar melakukan suatu tindakan yang mungkin secara prinsip sama tetapi secara proses berbeda.

POLY(many) dan MORPHISM(form). 
fungsinya sama isinya berbeda.
1. with function:
kita akan mencreate sebuah function yang bisa mengambil any object, mengikuti polymorphism.
create fungtionnya diluar class tersebut




contoh


```python
#contoh kodingan polymorphism
class cat():
    def sound(self):
        print("suara kucing = miaw")
    def walk(self):
        print("cara berjalan kucing = merangkak")
class dog():
    def sound(self):
        print("suara anjing = woofwoof")
    def walk(self):
        print("cara berjalan = merangkak")
class snake():
    def sound(self):
        print("suara ular = siasssttttttssssttttt")
    def walk(self):
        print("cara berjalan ular = melata")
def makesound(animal):
    animal.sound()
def makewalk(animal):
    animal.walk()
catObj=cat()
dogObj=dog()
snakeObj=snake()
makesound(catObj)
makewalk(catObj)
makesound(dogObj)
makewalk(dogObj)
makesound(snakeObj)
makewalk(snakeObj)
```

```python
class hewan():
    def __init__(self,nama):
        self.nama=nama
    def show(self):
        raise NotImplementedError("Subclass not implemented abstrack method")
class kucing(hewan):
    def show(self):
        return"kucing bersuara miaw dan berjalan dengan 4 kaki"
class anjing(hewan):
    def show(self):
        return"anjing bersuara woofwoof dan berjalan dengan 4 kaki"
class ular(hewan):
    def show(self):
        return"ular bersuara sttttstststststststs dan berjalan dengan melata"
hewan = [kucing("hewan1"),
        anjing("hewan2"),
        ular("hewan3")]
for hewan in hewan:
    print(hewan.nama+","+hewan.show())
```
